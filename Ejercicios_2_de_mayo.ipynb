```python
import random as rd

# Solicitar al usuario ingresar el número de filas y columnas para la matriz
m = int(input('Ingrese las filas:'))
n = int(input('Ingrese las columnas:'))

A = []  # Inicializando la matriz

# Insertar datos en la matriz
for i in range(m):
    fila = []
    for j in range(n):
        num = rd.randint(0, 10)  # Genera un número aleatorio entre 0 y 10
        fila.append(num)
    A.append(fila)

# Imprimir la matriz
for i in range(m):
    print('[', end='')
    for j in range(n):
        if j == n - 1:
            print(f'{A[i][j]}', end='')
        else:
            print(f'{A[i][j]}', end=',')
    print(']')

# Sumar una fila aleatoria
alea1 = rd.randint(0, m - 1)
suma1 = 0
for j in range(n):
    suma1 += A[alea1][j]
print(f'La suma de la fila {alea1} es {suma1}')

# Sumar una columna aleatoria
alea2 = rd.randint(0, n - 1)
suma2 = 0
for i in range(m):
    suma2 += A[i][alea2]
print(f'La suma de la columna {alea2} es {suma2}')

# Sumar la diagonal principal
suma_diagonal_prin = 0
if m == n:  # Verifica si la matriz es cuadrada
    for i in range(m):
        for j in range(n):
            if i == j:
                suma_diagonal_prin += A[i][j]
    print(f"La suma de la diagonal principal es: {suma_diagonal_prin}")
else:
    print("Solo podemos sumar la diagonal de matrices cuadradas")

# Sumar la diagonal secundaria
suma_diagonal_sec = 0
if m == n:  # Verifica si la matriz es cuadrada
    for i in range(m):
        for j in range(n):
            if i + j == n - 1:
                suma_diagonal_sec += A[i][j]
    print(f'La suma de la diagonal secundaria es {suma_diagonal_sec}')
else:
    print('No se puede sumar la diagonal de una matriz no cuadrada')

# Sumar las filas
suma_filas = []
for i in range(m):
    suma_fila = 0
    for j in range(n):
        suma_fila += A[i][j]
    suma_filas.append(suma_fila)
print(f"Las sumas de las filas son {suma_filas}")

# Sumar las columnas
suma_columnas = []
for i in range(n):
    suma_columna = 0
    for j in range(m):
        suma_columna += A[j][i]
    suma_columnas.append(suma_columna)
print(f"Las sumas de las columnas son {suma_columnas}")

# Encontrar el número mayor de cada fila y su índice
mayores_filas = []
for i in range(m):
    mayor_fila = A[i][0]
    indice = 0
    for j in range(n):
        if A[i][j] > mayor_fila:
            mayor_fila = A[i][j]
            indice = j
    mayores_filas.append((mayor_fila, indice))

print(f"Los números mayores de cada fila y sus índices son: {mayores_filas}")

# Encontrar el número mayor de cada columna y su índice
mayores_columnas = []
for i in range(n):
    mayor_columna = A[0][i]
    indice = 0
    for j in range(m):
        if A[j][i] > mayor_columna:
            mayor_columna = A[j][i]
            indice = j
    mayores_columnas.append((mayor_columna, indice))

print(f"Los números mayores de cada columna y sus índices son: {mayores_columnas}")
```
